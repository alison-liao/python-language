```python
# 674
#最长连续上升子序列长度
class Solution:
    def findLengthOfLCIS(self, nums) -> int:
        if (len(nums)== 0):
            return 0
        maxl=1
        currentl=1
        for i in range(1, len(nums)):
            if (nums[i]>nums[i-1]):
                currentl+=1
            else:
                currentl=1
            
            maxl=max(maxl, currentl)

        return maxl
```

```python
#516.最长子序列
class Solution:
    def longestPalindromeSubseq(self, s: str) -> int:

        n = len(s)
        dp = [[0]*n for i in range(n)]

        for i in range(n):
            dp[i][i] = 1

        for i in range(n,-1,-1):
            for j in range(i+1,n):
                if s[i] == s[j]:
                    dp[i][j] = dp[i+1][j-1] + 2
                else:
                    dp[i][j] = max(dp[i+1][j],dp[i][j-1])
        return dp[0][n-1]
```

```python
#198.相邻不选，最大多少
class Solution:
    def rob(self, nums) -> int:

        l=len(nums)
        if (not l):
            return 0
        value=[[0]*2 for i in range(l)]
        #print(value)

        value[0][1]=nums[0]
        maxv=nums[0]

        for i in range(1,l):
            value[i][1]=value[i-1][0]+nums[i]
            maxv=max(maxv, value[i][1])
            value[i][0]=max(value[i-1][0], value[i-1][1])
            maxv=max(maxv, value[i][0])
            
        return maxv
```

```python
#213
class Solution:
    def rob(self, nums) -> int:
        
        length = len(nums)
        if length == 0:
            return 0
        if length == 1:
            return nums[0]
        if length == 2:
            return max(nums[0], nums[1])
        
        def rob_action(nums):
            length = len(nums)
            dp = [0 for _ in range(length)]
            dp[0] = nums[0]
            dp[1] = max(nums[0], nums[1])
            for i in range(2, length):
                dp[i] = max(dp[i - 2] + nums[i], dp[i - 1])
            return dp
        
        nums1 = nums[1:]
        dp1 = rob_action(nums1)
        nums2 = nums[:-1]
        dp2 = rob_action(nums2)
        
        return max(dp1[-1], dp2[-1])
```
